```python
import pandas as pd
import glob
import os
import numpy as np

def normalize(column):
    return (column - column.min()) / (column.max() - column.min())
```

```python
path = os.getcwd()
df = pd.read_csv(path + "\\data\\Counties_Schools_Counts.csv")
display(df.head())
```

```python
df['Norm_Count'] = normalize(df['Total'])
display(df.head())
```

```python
df.to_csv(path + '\\data\\nc_schools_count.csv', index=False)
```

```python
path = os.getcwd()
df_clean = pd.read_csv(path + "\\data\\nc_schools_count.csv")
df_clean.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
fig, axs = plt.subplots(1, 4, sharey=True, figsize=(20, 5))
fig.suptitle('Distribution of Schools per County in North Carolina', fontsize=13, fontweight='bold')

axs[0].hist(df_clean['Pub_Count'], bins=30, color='blue', edgecolor='black', linewidth=0.3)
axs[0].set_title('Public Schools')
axs[0].set_xlabel('Number of Schools')

axs[1].hist(df_clean['Non_Pup_Count'], bins=30, color='orange', edgecolor='black', linewidth=0.3)
axs[1].set_title('Non-Public Schools')
axs[1].set_xlabel('Number of Schools')

axs[2].hist(df_clean['Uni_Count'], bins=30, color='green', edgecolor='black', linewidth=0.3)
axs[2].set_title('Universities')
axs[2].set_xlabel('Number of Schools')

axs[3].hist(df_clean['Norm_Count'], bins=50, color='red', edgecolor='black', linewidth=0.3)
axs[3].set_title('Normalized Score')
axs[3].set_xlabel('Score (0:worst - 1:best)')
```
